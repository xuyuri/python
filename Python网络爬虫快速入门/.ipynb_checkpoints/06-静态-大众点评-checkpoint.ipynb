# 实战
写爬虫流程：

1. 在浏览器中打开网站的某个url页面
2. 找网址规律（开发者工具）
3. 开发者工具定位
4. 保存数据
5. for循环把所有对url批量访问和保存

# 实战分类
- 静态网站html：如天涯、boss直聘、大众点评
- 动态网页json：京东评论、百度企业信用等

# 大众点评杭州站酒店


http://www.dianping.com/hangzhou/hotel/p2


```python
import csv
import requests 
from pyquery import PyQuery
import time

# mac
# csvf = open('output/大众点评酒店杭州.csv', 'a+', encoding='gbk', newline='')
csvf = open('output\\大众点评酒店杭州.csv', 'a+', encoding='gbk', newline='')
writer = csv.writer(csvf)
writer.writerow(('hotelname', 'place', 'price'))


headers = {'User-Agent': 'Mozilla/5.0 (Macintosh; Intel Mac OS X 10_14_6) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/77.0.3865.120 Safari/537.36'}

url_template = 'http://www.dianping.com/hangzhou/hotel/p{page}'
for page in range(1, 10):
    time.sleep(1)
    url = url_template.format(page=page)
    resp = requests.get(url, headers=headers)
    doc = PyQuery(resp.text)


    for item in doc.items('.content .hotelshop-list li .hotel-info-ctn'):
        name = item('.hotel-info-main .hotel-name a').text()
        place = item('.hotel-info-main .place').text()
        price = item('.hotel-remark .price p strong').text()
        writer.writerow((name, place, price))
        print(page, name, place, price)
    
csvf.close()
```

## 全国酒店爬虫

```python
import csv
import requests 
from pyquery import PyQuery
import time

# mac
# csvf = open('output/大众点评酒店杭州.csv', 'a+', encoding='gbk', newline='')
csvf = open('output\\大众点评酒店全国.csv', 'a+', encoding='gbk', newline='')
writer = csv.writer(csvf)
writer.writerow(('hotelname', 'place', 'price'))


headers = {'User-Agent': 'Mozilla/5.0 (Macintosh; Intel Mac OS X 10_14_6) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/77.0.3865.120 Safari/537.36'}

citys = ['chengdu', 'hangzhou', 'wuhan']
for city in citys:
    url_template = 'http://www.dianping.com/{city}/hotel/p{page}'
    for page in range(1, 3):
        #time.sleep(1)
        url = url_template.format(city=city, page=page)
        resp = requests.get(url, headers=headers)
        doc = PyQuery(resp.text)


        for item in doc.items('.content .hotelshop-list li .hotel-info-ctn'):
            name = item('.hotel-info-main .hotel-name a').text()
            place = item('.hotel-info-main .place').text()
            price = item('.hotel-remark .price p strong').text()
            writer.writerow((name, place, price))
            print(city, page, name, place, price)
    
csvf.close()
```
