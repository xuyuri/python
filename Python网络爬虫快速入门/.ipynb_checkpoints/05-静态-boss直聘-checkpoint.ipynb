# 实战
写爬虫流程：

1. 在浏览器中打开网站的某个url页面
2. 找网址规律（开发者工具）
3. 开发者工具定位
4. 保存数据
5. for循环把所有对url批量访问和保存

# 实战分类
- 静态网站html：如天涯、boss直聘、大众点评
- 动态网页json：京东评论、百度企业信用等

# boss直聘

https://www.zhipin.com/c101010100/?query=python&page=1&ka=page-1

![](img/boss.png)

```python
import requests
from pyquery import PyQuery
import csv
import time

#新建文件
# mac
# csvf = open('output/boss.csv', 'a+', encoding='gbk', newline='')
csvf = open('output\\boss.csv', 'a+', encoding='gbk', newline='')
writer = csv.writer(csvf)
writer.writerow(('work', 'salary'))

#伪装成浏览器，伪装成良民
headers = {'user-agent': 'Mozilla/5.0 (Macintosh; Intel Mac OS X 10_14_6) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/77.0.3865.120 Safari/537.36'}

cookies = {'cookie': '你自己的cookies'}

url_template = 'https://www.zhipin.com/c101010100/?query=python&page={p}&ka=page-{p}'


for page in range(1, 11):
    time.sleep(1)
    url = url_template.format(p=page)
    resp = requests.get(url, headers=headers, cookies=cookies)
    resp.encoding='utf-8'
    doc = PyQuery(resp.text)
    for item in doc.items('.job-primary .info-primary .name a'):
        work = item.find('.job-title').text()
        salary = item.find('.red').text()
        writer.writerow((work, salary))
        print(page, work, salary)

        
csvf.close()
```

这个网站让我们明白一个道理，爬虫要：
- 伪装成良民，用headers，cookies
- cookies有有效期
- 访问速度要慢
