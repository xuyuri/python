这种力度，这种中国红，让我想起2015年股灾。那场股灾真好，基本上大半国民都体验了一把股神的感觉。而我最为群众中的一份子，还是商学专业科班出身，不去炒股都不好意思出去溜达。铆足了劲儿，把暑假一个月挣到的两万多块钱转到海通证券账户里，一开始安全起见，我买了银行股，哎呀两万终于投进去了，YY着每天涨一两个点我就满足了，

>“黑子，今天我又赚了300，你要不也炒炒股.”
>“爸，你看，今天我又赚了500，比你们做生意强多了”。
>...

看到股市每天亮眼的表现，看到周围人每天涨幅都是7%、8%甚至是10%，而我安全的银行只有安稳的收益。离开学还有不到一个月，搏一搏吧，一天涨10%的话，7天就能翻翻，我不贪的。只要涨7天过后我就撤出，三年的研究生生活费就都有了。

终于我下定决心，说干咱就干。把银行股换成当时最火的题材股。当天就意外涨了10%，可惜天有不测风云，这感觉结果没开心几天，股市进入下行通道，每天暴跌10%，神州大地一片狼嚎，在这时候国家队挺身而出，当时我没有觉悟，还不是dang员，我没有跟国家队站在一起，而是从股市里撤出来了。结果又过了一阵，股市开始强势逆转，我错失回本的机会。

那段炒股的日子，每天茶饭不思，不思进取，只想着看大盘变红。我那两万块钱是读研前的暑假运营阿里巴巴挣到的钱，挣得很有成就感，也特别踏实。炒股呢，因为不懂，所以我只是股市中的小韭菜，又绿又长，等待着大佬们的收割。那时候我不知道进入股市是需要资本，需要技术，需要克服人性，当股市每天都涨的时候会觉得自己特牛掰，跟父母说


注意赚钱和挣钱不是一回事，挣钱是用力生钱，而挣钱是用钱生钱。挣钱不一定需要很多的本钱，智力劳动也可以挣钱，只需要有眼光和行动力就OK了，但是赚钱肯定是需要本钱，而且本钱必须要很大，除此之外还需要克制力，能在适当的时机智慧地不贪婪的主动退出，懂得明哲保身。

``踏踏实实挣钱很容易，但想躺着赚钱却很难。``
