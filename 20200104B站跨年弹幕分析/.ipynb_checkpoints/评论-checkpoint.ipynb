```python
import requests
import json
import csv

# 新建csv文件
csvf = open('评论.csv', 'a+', encoding='gbk', newline='')
writer = csv.writer(csvf)
writer.writerow(('User', 'Gender', 'Level', 'Message'))

#url构建
template = 'https://api.bilibili.com/x/v2/reply?callback=jQuery17204735001843722022_1578026445710&jsonp=jsonp&pn={page}&type=1&oid=81403056&sort=2&_=1578028992809'

headers = {'User-Agent': 'Mozilla/5.0 (Macintosh; Intel Mac OS X 10_15_1) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/79.0.3945.88 Safari/537.36',
           'Referer': 'https://www.bilibili.com/bangumi/play/ep307609'}

# 访问，保存
for page in range(1, 10):
    url = template.format(page=page)
    resp = requests.get(url, headers=headers)
    data = json.loads(resp.text[41:-1])['data']
    replies = data['replies']
    for reply in replies:
        message = reply['content']['message']
        name = reply['member']['uname']
        gender = reply['member']['sex']
        level = reply['member']['level_info']['current_level']
        print(message, name)
        try:
            writer.writerow((name, gender, level, message))
        except:
            pass
        
# 关闭csv
csvf.close()
```
