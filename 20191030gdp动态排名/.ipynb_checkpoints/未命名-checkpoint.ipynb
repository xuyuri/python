### 颜色标签

```python
import random

colors = ['#adb0ff', '#ffb3ff', '#90d595', '#e48381', '#aafbff', '#f7bb5f', '#eafb50', 
         '#adb0ff', '#ffb3ff', '#90d595', '#e48381', '#aafbff', '#f7bb5f', '#eafb50',
         '#adb0ff', '#ffb3ff', '#90d595', '#e48381', '#aafbff', '#f7bb5f', '#eafb50',
         '#adb0ff', '#ffb3ff', '#90d595', '#e48381', '#aafbff', '#f7bb5f', '#eafb50',
         '#adb0ff', '#ffb3ff', '#90d595', '#e48381', '#aafbff', '#f7bb5f', '#eafb50',
         '#adb0ff', '#ffb3ff']

#给每个城市随机分配颜色
citycolors = dict()
citys = set(df['地区'])
for color, city in zip(colors, citys):
    citycolors[city] = color
    
citycolors
```

```python
%matplotlib inline


fig, ax = plt.subplots(figsize=(15, 8))
#前7个城市
ax.barh(df2['地区'], df2['2001年'], color=[citycolors[city] for city in df2['地区']])
ax.text(1, 0.45, '2001年', transform=ax.transAxes, size=46, ha='right')
```

### 细节刻画

为了方便起见，我们将代码迁移到draw_barchart函数中：



- Text：更新字体大小、颜色、方向
- Axis：将x轴移动到顶部，添加颜色和副标题
- Grid：在bars添加
- Format：逗号分隔值和轴标记符
- Remove：框架，y轴标签



```python
fig, ax = plt.subplots(figsize=(15, 8))

def draw_barchart(year):
    year = str(year)+'年'
    df = pd.read_excel('主要城市年度数据.xls', skiprows=[0, 1, 2])
    df2 = df[['地区', year]].sort_values(by=year, ascending=False).head(7)
    
    ax.clear()
    ax.barh(df2['地区'], df2[year], color=[citycolors[city] for city in df2['地区']])
    dx = df[year].max()/200
        
    ax.text(1, 0.45, year, transform=ax.transAxes, color='#777777', size=46, ha='right', weight=800)
    ax.text(0, 1.06, 'GDP(单位:亿元)', transform=ax.transAxes, size=12, color='#777777')
    #ax.xaxis.set_major_formatter(ticker.StrMethodFormatter('{x:,.0f}'))
    ax.xaxis.set_ticks_position('top')
    ax.tick_params(axis='x', colors='#777777', labelsize=12)
    #ax.set_yticks([])
    ax.margins(0, 0.01)
    ax.grid(which='major', axis='x', linestyle='-')
    ax.set_axisbelow(True)
    ax.text(0.4, 1.12, '2000~2019中国城市GDP变迁',
           transform=ax.transAxes, size=24, weight=600, ha='left')
    
    plt.box(False)
    
 
draw_barchart(2017)
```

### 引入动画

```python
import matplotlib.animation as animation
from IPython.display import HTML

fig, ax = plt.subplots(figsize=(15, 8))
animator = animation.FuncAnimation(fig, draw_barchart, frames=range(2000, 2019))
HTML(animator.to_jshtml())
```
